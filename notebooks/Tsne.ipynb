```python
from vega import VegaLite
from IPython.display import display
import json
import csv
```

```python
def generate_scatterplot_with_text(data, x="x", y="y", size=1500):

    scatterplot= {
        "width":size,
        "height":size,
        "data": data,
        "layer":
        [
            {
             "mark":"circle",   
             "encoding":
                {
                "x":{ "field": x, "type": "quantitative"},
                "y":{ "field": y, "type": "quantitative"},
                "color":{ "field":"color", "type":"nominal"}
                }
            },
            {
            "mark":
                {
                "type":"text",
                "align":"center",
                "baseline":"bottom",
                "dy":-10
                },
            "encoding":
                {
                "x":{ "field": x, "type": "quantitative"},
                "y":{ "field": y, "type": "quantitative"},
                "text":{ "field":"media", "type":"nominal"},
                "color":{ "field":"color", "type":"nominal"}
                }
            }
        ]
        }
              
 
  
    return scatterplot
```

```python
with open('data/reg_dim_media_tsne.json', 'r') as f:
    values=json.load(f)
    
values = {"values":values}

scatterplot = generate_scatterplot_with_text(values)

VegaLite(scatterplot)
```
